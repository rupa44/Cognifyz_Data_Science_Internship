LEVEL 3 TASK 1

Task: Predictive Modeling
1.Build a regression model to predict theaggregate rating of a restaurant based on available features.Split the dataset into training and testing sets
and evaluate the model's performance using appropriate metrics.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
df1=pd.get_dummies(df,columns=["Has Table booking","Has Online delivery"],drop_first=True)

```

```python
x=df1[["Average Cost for two","Votes","Price range","Has Table booking_Yes","Has Online delivery_Yes"]]
y=df1["Aggregate rating"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("model: Linear regression")
print("Mean Squared Error:",mse)
print("R-squared:",r2)
```

2.Experiment with different algorithms (e.g.,linear regression, decision trees,random forest) and compare their performance

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
models={
    "Linear Regression" : LinearRegression(),
    "Decision Tree Regressor" : DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor" : RandomForestRegressor(random_state=42)
}


result = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    result[name] = {
        "Mse": mse,
        "r2 score": r2
    }

results = pd.DataFrame(result).T  # Transpose for better readability
print(results)
```

Task: Customer Preference Analysis
1.Analyze the relationship between the type of cuisine and the restaurant's rating.

```python
cuisines=df1["Cuisines"]
cuisines.value_counts().head(10)
top_cuisines=cuisines.value_counts().head(10).index
cuisine_rate=pd.DataFrame({"cuisine": cuisines,"Rating":df1["Aggregate rating"]})
cuisine_rate_top=cuisine_rate[cuisine_rate["cuisine"].isin(top_cuisines)]
plt.figure(figsize=(10,6))
sns.boxplot(x="cuisine",y="Rating",data=cuisine_rate_top,palette = "viridis")
plt.title("Relationship  between cuisines and aggregate rating")
plt.xlabel("Cuisines")
plt.ylabel("Aggregate Rating")
plt.xticks(rotation=45)
plt.show()
```

2.Identify the most popular cuisines among customers based on the number of votes.

```python
cuisines_vote=pd.DataFrame({"Cuisine":cuisines,"Votes" :df1["Votes"]})
cuisines_vote_sum=cuisines_vote.groupby("Cuisine")["Votes"].sum()
pop_cusines=cuisines_vote_sum.sort_values(ascending=False)
print("top 10 popular cuisines based on votes:")
print(pop_cusines.head(10))
```

```python
plt.figure(figsize=(10,6))
pop_cusines.head(10).plot(kind="bar",color="skyblue")
plt.title("TOP 10 POPULAR CUISINES")
plt.xlabel("CUISINES")
plt.ylabel("NO OF VOTES")
plt.show()
```

3.Determine if there are any specific cuisines that tend to receive higher ratings.

```python
cuisines_rate=pd.DataFrame({"Cuisine":cuisines,"Rating":df1["Aggregate rating"]})
average_rate=cuisines_rate.groupby("Cuisine")["Rating"].mean()
sort_cus=average_rate.sort_values(ascending=False)
print("top 10 cuisines with highest average ratings")
print(sort_cus.head(10))
```

```python
plt.figure(figsize=(8,6))
sort_cus.head(10).plot(kind="barh",color="skyblue")
plt.title("TOP 10 HIGH AVERAGE RATINGS")
plt.xlabel("AVERAGE RATING")
plt.ylabel("CUISINE")
plt.show()
```
